### Some data clean up that's necessary for our project

```python
import pandas as pd
import numpy as np
```

```python
spotify = pd.read_csv('/Users/michellehewson/Desktop/Harvard/CS171/FinalProject/cs171_finalproject/data/spotify_top_charts_22.csv')  
tiktok = pd.read_csv('/Users/michellehewson/Desktop/Harvard/CS171/FinalProject/cs171_finalproject/data/TikTok_songs_2022 2.csv')
```

```python
# Checking out our data
spotify.head()
```

```python
# Need to clean up the names of tracks (the syntax between both of the datasets is not the same)
# TikTok dataset has a lot of "SONG (feat. artist2, artist3)"
# Spotify dataset has "SONG" and then the artist names in the artist_names column
# For simplicity's sake, we'll just count the first artist listed as the owner of the song and ignore the features

spotify['old_track_name'] = spotify['track_name']
spotify['track_name'] = spotify['track_name'].str.split('(', n=1).str[0]
spotify['track_name'] = spotify['track_name'].str.split('-', n=1).str[0]

tiktok['old_track_name'] = tiktok['track_name']
tiktok['track_name'] = tiktok['track_name'].str.split('(', n=1).str[0]
tiktok['track_name'] = tiktok['track_name'].str.split('-', n=1).str[0]
```

```python
spotify['old_artist_names'] = spotify['artist_names']
spotify['artist_names'] = spotify['artist_names'].str.split(',', n=1).str[0]
spotify = spotify.rename(columns={'artist_names': 'artist_name'})
spotify.head()
```

```python
# all of the songs that appear in both datasets (so they're a top song on tiktok AND spotify)
merged_df = pd.merge(tiktok, spotify, on='track_name', how='inner')
```

```python
# make some new CSVs
#spotify.to_csv('spotify_clean.csv', index=False)
#tiktok.to_csv('tiktok_clean.csv', index=False)
#merged_df.to_csv('tiktok_spotify_merged.csv', index=False)
```

```python
# Counting how many songs an artist has (they're the first one listed) in the top spotify charts
spotify_artist_counts = spotify['artist_name'].value_counts().reset_index()
spotify_artist_counts.columns = ['artist', 'count']

spotify_artist_counts = pd.DataFrame(spotify_artist_counts)
spotify_artist_counts.head()
```

```python
tiktok_artist_counts = tiktok['artist_name'].value_counts().reset_index()
tiktok_artist_counts.columns = ['artist', 'count']
tiktok_artist_counts.head()
```

```python
#spotify_artist_counts.to_csv('spotify_artist_counts.csv', index=False)
#tiktok_artist_counts.to_csv('tiktok_artist_counts.csv', index=False)
```

```python
# creating specific dataframe for key information in the tiktok and spotify datasets
```

```python
num_tiktok_songs = len(tiktok)
num_spotify_songs = len(spotify)
```

```python
tiktok_keys = tiktok['key'].value_counts()
tiktok_keys = pd.DataFrame(tiktok_keys).reset_index()
tiktok_keys['percentage'] = round((tiktok_keys['count'] / num_tiktok_songs),3) * 100

tiktok_keys
```

```python
spotify_keys = spotify['key'].value_counts()
spotify_keys = pd.DataFrame(spotify_keys).reset_index()
spotify_keys['percentage'] = round((spotify_keys['count'] / num_spotify_songs),3) * 100

spotify_keys
```
