### Some data clean up that's necessary for our project

```python
import pandas as pd
import numpy as np
```

```python
spotify = pd.read_csv('/Users/michellehewson/Desktop/Harvard/CS171/FinalProject/cs171_finalproject/data/spotify_top_charts_22.csv')  
tiktok = pd.read_csv('/Users/michellehewson/Desktop/Harvard/CS171/FinalProject/cs171_finalproject/data/TikTok_songs_2022 2.csv')
```

```python
# Checking out our data
spotify.head()
```

```python
# For the spotify data, we just want the first artist in the artist_names column when
# we count how many songs an artist has in the top spotify charts (for simplicity)
spotify['first_artist'] = spotify['artist_names'].str.split(',').str[0]

spotify_artist_counts = spotify['first_artist'].value_counts().reset_index()
spotify_artist_counts.columns = ['artist', 'count']

spotify_artist_counts = pd.DataFrame(spotify_artist_counts)
```

```python
spotify_artist_counts.head()
```

```python
tiktok.head()
tiktok_artist_counts = tiktok['artist_name'].value_counts().reset_index()
tiktok_artist_counts.columns = ['artist', 'count']
```

```python
tiktok_artist_counts.head(5)
```

```python
spotify_artist_counts.to_csv('spotify_artist_counts.csv', index=False)
tiktok_artist_counts.to_csv('tiktok_artist_counts.csv', index=False)
```
